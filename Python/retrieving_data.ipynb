# This is the code for cleaning the data

```python
# Dependencies
import pandas as pd
 
# Store filepath in a variable
crimes_csv = "Data/Crimes_2023.csv"

# Read the csv
crimes_df = pd.read_csv(crimes_csv, encoding="ISO-8859-1")

# Dropping nas
crimes_df.dropna(inplace = True)

# Print the head
crimes_df.head(10)
```

```python
# Columns
crimes_df.info()
```

```python
# Cleaning columns out
labels = ["ï»¿ID", "Case Number", "Block", "IUCR", "Updated On"]
crimes_df.drop(labels, axis=1, inplace=True)
crimes_df.head(10)
```

```python
# Dividing date column
date_df = crimes_df["Date"].str.split(" ", n = 1, expand = True)

# Inserting new columns in data frame
crimes_df["Dates"] = date_df[0]
crimes_df["Time"] = date_df[1]

# Dropping Date column
crimes_df.drop(columns =["Date"], inplace = True)

# Show df
crimes_df.head(10)
```

```python
# Checking for nas
crimes_df.info()
```

```python
crimes_df["Month"] = crimes_df['Dates'].astype('datetime64').dt.strftime("%B")
```

```python
# Export clean csv  
crimes_df.to_csv("Data/Clear_Crimes_2023.csv", index=True)
```
